## Описание задачи

Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться.  
Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов.   


Необходимо разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.  

1. Нужно промаркировать уровень финансовой активности постоянных покупателей. В компании принято выделять два уровня активности: «снизилась», если клиент стал покупать меньше товаров, и «прежний уровень».   


2. Нужно собрать данные по клиентам по следующим группам:  
 - Признаки, которые описывают коммуникацию сотрудников компании с клиентом.   
 - Признаки, которые описывают продуктовое поведение покупателя. Например, какие товары покупает и как часто.  
 - Признаки, которые описывают покупательское поведение клиента. Например, сколько тратил в магазине.  
 - Признаки, которые описывают поведение покупателя на сайте. Например, как много страниц просматривает и сколько времени проводит на сайте.  


3. Нужно построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца.  
4. В исследование нужно включить дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца.   
5. Используя данные модели и данные о прибыльности клиентов, нужно выделить сегменты покупателей и разработать для них персонализированные предложения

```python
def find_duplicated(dataframe):
    """ A function for duplicates search and counting"""
    
    duplicates = dataframe.duplicated().sum()
    
    duplicates_percentage = (duplicates / len(dataframe)) * 100
    
    duplicates_summary = pd.DataFrame({
        'Количество дубликатов': duplicates,
        'Процент дубликатов': duplicates_percentage
    },  index=[0])
    duplicates_summary = duplicates_summary[duplicates_summary['Количество дубликатов'] > 0]
    duplicates_summary = duplicates_summary.sort_values(by='Количество дубликатов', ascending=False)
    
    return duplicates_summary, dataframe[dataframe.duplicated()]

find_duplicated(market)
find_duplicated(market_money)
find_duplicated(market_time)
find_duplicated(money)
```

## Использование пайплайнов

### Создание пайплайнов подготовки данных

Разделим данные на обучающую и тестовую выборки.

```python
df = merged_df.drop('id', axis=1)
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   df.drop(['Покупательская_активность'], axis=1),
   df['Покупательская_активность'],
   test_size=TEST_SIZE,
   random_state=RANDOM_STATE,
   stratify = df['Покупательская_активность'])
```

```python
# Введем обозначения для типов исходных данных
# Категориальные признаки для кодирования OneHotEncoder
ohe_columns = [
    'Тип_сервиса',
    'Разрешить_сообщать',
    'Популярная_категория'
]

# Количественные признаки без выручки за препредыдущий месяц (Выручка_1_месяц)
numeric_columns = [
    'Маркет_актив_6_мес',
    'Маркет_актив_тек_мес',
    'Длительность',
    'Акционные_покупки',
    'Средний_просмотр_категорий_за_визит',
    'Неоплаченные_продукты_штук_квартал',
    'Ошибка_сервиса',
    'Страниц_за_визит',
    'Выручка_1_месяц',
    'Выручка_2_месяц',
    'Выручка_3_месяц',
    'Время_2_месяц',
    'Время_3_месяц'
]
```

```python
# В исходных данных пропусков нет, однако, это может измениться в будущем, поэтому сразу добавим обработку.
ohe_pipe = Pipeline(
    [
        (
        'simpleImputer_ohe',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        ('ohe',
        OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)
```

```python
# Создаем общий пайплайн подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('num', MinMaxScaler(), numeric_columns)
    ],
    remainder='passthrough'
)
```

В результате созданы пайпланы для обработки данных различного типа:
 - Категориальные признаки для кодирования OneHotEncoder подготавливались в пайплайне ohe_pipe
Общий пайплайн подготовки данных включает в себя обработку категориальных признаков и количественных с нормализацией MinMaxScaler.

### Общий пайплайн и перебор моделей

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
param_distributions = [
    # словарь для модели KNeighborsClassifier
    {
    'models': [KNeighborsClassifier()],
    'models__n_neighbors': range(1,10),
    'preprocessor__num': [StandardScaler(), MinMaxScaler()]
    },
    # словарь для модели DecisionTreeClassifier
    {
    'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
    'models__max_depth': range(2,10),
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    # словарь для модели SVC
    {
    'models':[SVC(random_state=RANDOM_STATE,
                 probability=True)],
    'models__kernel': ['rbf', 'linear', 'sigmoid'],
    'models__C': [0.1, 1, 10, 100],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    # словарь для модели LogisticRegression
    {
    'models': [LogisticRegression(random_state=RANDOM_STATE, solver='saga', max_iter=1000)],
    'models__C': [0.01, 0.1, 1, 3],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
```

```python
warnings.filterwarnings("ignore", category=RuntimeWarning)
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_distributions,
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)
```

```python
# Присваиваем всем id обратно
ids = merged_df['id'].tolist()
df['id'] = ids

# Создание DataFrame с вероятностями
df_prob = pd.DataFrame({
    'id': df['id'],  # Предполагается, что у вас есть 'id' в df
    'y_probability': y_probability
})

# Объединение DataFrame по 'id'
merged_df = merged_df.merge(df_prob, on='id', how='left')

# Создаем два сегмента на основе данных
# Определяем исследуемый сегмент (высокая вероятность снижения и высокая прибыль)
seg_1 = merged_df[(y_probability >= 0.7) & (merged_df['Прибыль'] >= 3)]

# Остальные сегменты
other_segments = merged_df[(y_probability < 0.7) | (merged_df['Прибыль'] < 3)]

# Построение scatter plot для двух сегментов
plt.figure(figsize=(12, 6))
sns.scatterplot(data=seg_1, x='y_probability', y='Прибыль', label='Исследуемый сегмент', color='orange')
sns.scatterplot(data=other_segments, x='y_probability', y='Прибыль', label='Другие сегменты', color='blue')

# Линии порогов для вероятности снижения и прибыли
plt.axvline(x=0.7, color='green', linestyle='--')
plt.axhline(y=3, color='green', linestyle='--')

plt.legend()
plt.show()

```
