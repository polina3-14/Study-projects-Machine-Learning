```python
def find_duplicated(dataframe):
    """ A function for duplicates search and counting"""
    
    duplicates = dataframe.duplicated().sum()
    
    duplicates_percentage = (duplicates / len(dataframe)) * 100
    
    duplicates_summary = pd.DataFrame({
        'Количество дубликатов': duplicates,
        'Процент дубликатов': duplicates_percentage
    },  index=[0])
    duplicates_summary = duplicates_summary[duplicates_summary['Количество дубликатов'] > 0]
    duplicates_summary = duplicates_summary.sort_values(by='Количество дубликатов', ascending=False)
    
    return duplicates_summary, dataframe[dataframe.duplicated()]
display(find_duplicated(train_job_sat))
display(find_duplicated(test_merged))
```

```python
# Используем функцию для исследовательского анализа количественных признаков
def describe_numeric(dataframe, param, bins=20, color='violet'):
    """ A function for exploratory analysis of numeric parameters"""
    
    # Описательная статистика
    print(dataframe[param].describe())
    
    # Гистограмма
    plt.figure(figsize=(10, 5))
    plt.hist(dataframe[param], bins=bins, color='dodgerblue', edgecolor='black')
    plt.title(f'Гистограмма распределения параметра {param} из таблицы {dataframe.name}')
    plt.xlabel(f'{param}')
    plt.ylabel('Частота')
    plt.show()
    
    # Ящик с усами
    plt.figure(figsize=(10, 5))
    dataframe[param].plot.box(title=f'Ящик с усами для {param} из таблицы {dataframe.name}')
    plt.show()
    
for table in table_list:
    for parameter in num_columns:
        describe_numeric(table, parameter)
```

```python
# Воспользуемся функцией для исследовательского анализа категориальных признаков.
def pie_plot(dataframe, param):
    plt.figure(figsize=(10, 6))
    dataframe[param].value_counts().plot.pie(title=f'Круговая диаграмма {param} из таблицы {dataframe.name}',
                                      autopct='%1.1f%%', 
                                      startangle=90,
                                      wedgeprops={'edgecolor': 'black'},
                                      labels=dataframe[param].unique())
    plt.ylabel('')
    plt.show()

for table in table_list:
    for parameter in cat_columns:
        pie_plot(table, parameter)
```

```python
# Введем обозначения для типов исходных данных
# Категориальные признаки для кодирования OneHotEncoder
ohe_columns = [
    'dept',
    'last_year_promo',
    'last_year_violations'
]

# Категориальные признаки для кодирования OrdinalEncoder
ord_columns = [
    'level',
    'workload'
]

# Количественные признаки 
numeric_columns = [
    'employment_years',
    'salary'
]
```

В исходных данных есть пропуски, поэтому обработаем их SimpleImputer. Обработаем разные типы данных в отдельных пайплайнах, которые объединим в общий пайплайн подготовки данных.

```python
# Заменяем пробелы на NaN



# Создаём пайплайн для подготовки признаков из списка ohe_columns
ohe_pipe = Pipeline(
    [
        (
        'simpleImputer_before_ohe',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        ('ohe',
        OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)

# Создаём пайплайн для подготовки признаков из списка ord_columns
ord_pipe = Pipeline(
    [
        (
        'simpleImputer_before_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ord',
        OrdinalEncoder(
            categories=[
                ['junior', 'middle', 'senior'],
                ['low', 'medium', 'hard']
            ],
            handle_unknown='use_encoded_value', unknown_value=np.nan)
        ),
        ('simpleImputer_after_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )              
    ]
)

# Cоздаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), numeric_columns)
    ], 
    remainder='passthrough'
)
```

### Обучение моделей

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
])

param_grid = [
    # словарь для модели DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_features': range(1,6),
        'models__min_samples_leaf': range(1,4),
        'models__min_samples_split':range(1,16),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # словарь для модели LinearRegression()
    
    {
        'models': [LinearRegression()],
        'preprocessor__num': [MinMaxScaler(), StandardScaler()]
    }
]
```

Используем метрику SMAPE (англ. symmetric mean absolute percentage error, «симметричное среднее абсолютное процентное отклонение») для определения относительной ошибки предсказания в задаче регрессии. Метрика позволяет сделать все объекты в выборках равнозначными. Значение 0 означает отсутствие ошибок в предсказаниях значений.

```python
def smape(y_true, y_pred):
    """
    Функция для вычисления SMAPE (Symmetric Mean Absolute Percentage Error).
    
    """
    
    # Преобразуем данные в numpy массивы
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    # Вычисляем абсолютные разницы между истинными и предсказанными значениями
    numerator = np.abs(y_true - y_pred)
    
    # Вычисляем среднее значение между истинными и предсказанными значениями
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    
    # Вычисляем SMAPE
    smape_value = np.mean(numerator / denominator) * 100
    
    return smape_value
```

```python
# Используем SMAPE в качестве метрики
smape_scorer = make_scorer(smape, greater_is_better=False)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
warnings.filterwarnings("ignore", category=RuntimeWarning)
# Используем рандомизированный поиск для подбора лучшей модели по метрике SMAPE
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring=smape_scorer,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)


print('Лучшие параметры модели:\n', randomized_search.best_params_)
print('Лучшая метрика модели на тренировочных данных:\n', (-1) * randomized_search.best_score_)
```

```python
# Получение предсказаний на тестовой выборке
y_pred = randomized_search.best_estimator_.predict(X_test)
# Вычисление SMAPE на тестовой выборке
smape_test = smape(y_test, y_pred)

print('SMAPE на тестовой выборке:', smape_test)
```

В результате рандомизированного поиска выявлена лучшая модель -  DecisionTreeRegressor со следующими параметрами:
- min_samples_split=6 (минимум объектов в узле для разделения),
- min_samples_leaf=3 (минимальное количество объектов в одном листе),
- max_features=5 (максимум признаков, которые контролируются при поиске решения).  

Метрика SMAPE для тестовой выборки 14.3%, что может говорить о небольшой ошибке предсказаний. 

Модель линейной регрессии, скорее всего, слишком простая для определения уровня удовлетворенности работой и не учитывает так много факторов, как модель решающих деревьев.

Анализ частоты увольнений по отделам, уровню должности и уровню загруженности

```python
def analyze_quit_rate(param):
    """
    Функция для анализа частоты увольнений по различным параметрам
    
    """
    param_rate = (
        train.groupby(param)['quit']
        .value_counts(normalize=True)
        .rename("quit_rate")
        .reset_index()
    )
    
    quit_rate_data = param_rate.loc[param_rate['quit'] == True]
    
    print(f'Частота увольнений по {param}:\n', quit_rate_data)

    plt.figure(figsize=(10, 6))
    sns.barplot(data=quit_rate_data, x=param, y='quit_rate')
    plt.title(f'Частота увольнений по {param}')
    plt.ylabel("Частота увольнений")
    plt.xlabel(param)
    plt.show()
```

### Подготовка данных

Подготовьте признаки так же, как и в первой задаче: выполните подготовку в пайплайне, дополнив пайплайн предобработки. При кодировании учитывайте особенности признаков и моделей и используйте как минимум два кодировщика.

```python
X_train = train.drop(['id', 'quit'], axis=1)
y_train = train['quit']
X_test = test.drop(['id','quit'], axis=1)
y_test = test['quit']
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

Подготовим данные для кодирования

```python
# Введем обозначения для типов исходных данных
# Категориальные признаки для кодирования OneHotEncoder
ohe_columns = [
    'dept',
    'last_year_promo',
    'last_year_violations'
]

# Категориальные признаки для кодирования OrdinalEncoder
ord_columns = [
    'level',
    'workload'
]

# Количественные признаки 
numeric_columns = [
    'employment_years',
    'salary',
    'job_satisfaction_rate_pred'
]
```

```python

# пайплайн для подготовки признаков из списка ohe_columns
ohe_pipe = Pipeline(
    [
        (
        'simpleImputer_before_ohe',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        ('ohe',
        OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)

# пайплайн для подготовки признаков из списка ord_columns
ord_pipe = Pipeline(
    [
        (
        'simpleImputer_before_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ord',
        OrdinalEncoder(
            categories=[
                ['junior', 'middle', 'senior'],
                ['low', 'medium', 'hard'],
            ],
            handle_unknown='use_encoded_value', unknown_value=np.nan)
        ),
        ('simpleImputer_after_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )              
    ]
)

# общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), numeric_columns)
    ], 
    remainder='passthrough'
)
```

### Обучение модели

```python
# пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

parameters_grid = [
    # словарь для модели KNeighborsClassifier
    {
    'models': [KNeighborsClassifier()],
    'models__n_neighbors': range(1,15),
    'preprocessor__num': [StandardScaler(), MinMaxScaler()]
    },
    # словарь для модели DecisionTreeClassifier
    {
    'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
    'models__max_features': range(2,6),
    'models__min_samples_leaf': range(2,4),
    'models__min_samples_split':range(2,16),
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    # словарь для модели SVC
    {
    'models':[SVC(random_state=RANDOM_STATE,
                 probability=True)],
    'models__kernel': ['rbf', 'linear'],
    'models__C': [0.1, 1, 10, 100],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    # словарь для модели LogisticRegression
    {
    'models': [LogisticRegression(random_state=RANDOM_STATE)],
    'models__C': [0.1,0.4,0.5,0.6,0.7,0.8,1,2,5,10],
    'preprocessor__num': [StandardScaler(), MinMaxScaler()]
    }
]
```

```python
warnings.filterwarnings("ignore", category=RuntimeWarning)
# Используем рандомизированный поиск для подбора лучшей модели по метрике ROC-AUC
randomized_search = RandomizedSearchCV(
    pipe_final,
    parameters_grid,
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)


print('Лучшие параметры модели:\n\n', randomized_search.best_params_)
print('Лучшая метрика модели на тренировочных данных:\n', randomized_search.best_score_)
```

```python
y_pred_proba = randomized_search.best_estimator_.predict_proba(X_test)[:,1]
roc_auc = roc_auc_score(y_test, y_pred_proba)
print('Метрика лучшей модели на тестовой выборке (ROC AUC):', roc_auc)
```

Метрика лучшей модели на тестовой выборке получилась ниже среднего - модель переобучилась. Попробуем произвести отбор признаков. Также выведем лучшие модели по метрике ROC-AUC и их параметры.

```python
results = randomized_search.cv_results_

pd.set_option('display.max_colwidth', None) 
pd.set_option('display.expand_frame_repr', False) 
results_df = pd.DataFrame(results)
top_models = results_df.sort_values(by='mean_test_score', ascending=False).head(10)

# Вывод топ-10 моделей
display(top_models[['params', 'mean_test_score']])
```

```python
# пайплайн: подготовка данных и модель с отбором признаков
pipe_final_selection = Pipeline([
    ('preprocessor', data_preprocessor),
    ('feature_selection', SelectKBest(score_func=mutual_info_classif,k=8)),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

warnings.filterwarnings("ignore", category=RuntimeWarning)
# Используем рандомизированный поиск для подбора лучшей модели по ROC-AUC c отбором признаков
randomized_search_selection = RandomizedSearchCV(
    pipe_final_selection,
    parameters_grid,
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search_selection.fit(X_train, y_train)


print('Лучшие параметры модели:\n\n', randomized_search_selection.best_params_)
print('Лучшая метрика модели на тренировочных данных:\n', randomized_search_selection.best_score_)
```

```python
y_pred_proba_s = randomized_search_selection.best_estimator_.predict_proba(X_test)[:,1]
roc_auc = roc_auc_score(y_test, y_pred_proba_s)
print('Метрика лучшей модели на тестовой выборке (ROC AUC):', roc_auc)
```

На целевой признак - уровень удовлетворенности работой сильнее всего влияют:
- Оценка руководителя (0.76-0.77) - предположительно, при высокой оценке руководителя оценка сотрудника выше,

Для второй задачи тестовые данные совпадают с данными первой задачи.

Количественные признаки распределены так же, как и для первой задачи:

- Длительность трудоустройства в годах описывается распределением Пуассона, в компании преобладают недавно устроившиеся сотрудники. В среднем, на момент исследования, сотрудники работают около 3 лет.

- Зарплата стремится к нормальному распределению, есть установленная граница минимума 12 тыс., существует также небольшая доля сотрудников, зарабатывающих больше других. Медианная зп - 30 тыс рублей.  

Описание категориальных признаков также похоже на данные для первой задачи:
- Большая часть сотрудников из тренировочной выборки работают в продажах, тестовых данных - в маркетинге.
- Почти половина сотрудников - middle-специалисты (48.7%), затем 42.4-42.% junior, senior меньше всего. Этот признак можно закодировать Ordinal Encoder.  
- Чуть больше половины сотрудников (53%) в тренировочной выборке имеют высокую загруженность рабочими задачами, 30.2% - средне. В тестовой выборке среднюю загруженность имеют 52.1% сотрудников, высокую нагрузку только 18.2%.   Этот признак можно закодировать Ordinal Encoder.  
- Только 2.8-3.2% сотрудников получили повышение в прошлом году.  
- Почти четверть сотрудников в прошлом году нарушали трудовой договор (13.6-13.1).  
- 28.2% сотрудников уволились - целевой признак представлен одинаково в обеих выборках.  
Данные представлены достаточно равномерно, за исключением рабочей нагрузки.

Мультиколлинеарности между входными признаками нет.   

Высокий коэффициент корреляции Фи наблюдается между:  
- Рабочей нагрузкой и зарплатой (0.74-0.79),
- Уровнем должности и зарплатой (0.75),
- Уровнем должности и длительностью трудоустройства (0.69),
- Длительностью трудоустройства и зарплатой (0.49).

На целевой признак - увольнение - сильнее всего влияют:
- Длительность трудоустройства (0.61-0.66) - скорее всего, сотрудник принимает решение развиваться дальше, ему не хватает карьерных возможностей в данной компании,
- Зарплата (0.51-0.56) - скорее всего, сотрудники уходят за более высокой зарплатой.

В рамках решения второй задачи составлен портрет уволившегося сотрудника.

**Портрет уволившегося сотрудника:**
- Отдел: Чаще всего увольняются сотрудники из отдела технологий.
- Уровень должности: Младший уровень должности больше подвержен увольнению.
- Стаж работы: Небольшой стаж работы больше подвержен увольнению.
- Средняя зарплата ниже у уволившегося сотрудника.
- Уровень загруженности: низкий уровень загрузки у уволившихся сотрудников.
- Последнее повышение: сотрудники, не получавшие повышение в прошлом году, чаще покидают компанию.
- Нарушения: чаще увольняютсяя сотрудники, нарушавшие трудовой договор.
- Оценка работы: примерно одинакова для сотрудников уволившихся и оставшихся.
- Зарплата: Средняя зарплата уволившихся ниже, чем у оставшихся в компании.

Также проверена связь между увольнениями и удовлетворенностью сотрудников работой. Выявлено, что сотрудники, менее довольные работой (с оценкой ниже 0.5) чаще увольняются, чем сотрудники, чья оценка работы более 0.5. Решено добавить признак удовлетворенности работой к входным признакам для обучения моделей.

** 5. Обучение моделей. **

**Для первой задачи определения уровня удовлетворенности сотрудников:**

Подготовка данных и обучение лучшей модели реализовано в pipe_final. Гиперпараметры моделей деревьев решений и линейной регрессии указаны в словаре param_grid. 
Для оценки качества моделей использована метрика SMAPE (англ. symmetric mean absolute percentage error, «симметричное среднее абсолютное процентное отклонение») для определения относительной ошибки предсказания в задаче регрессии. Метрика позволяет сделать все объекты в выборках равнозначными.  
Подбор лучшей модели осуществлялся с помощью рандомизированного поиска.
В результате рандомизированного поиска выявлена лучшая модель - **DecisionTreeRegressor** со следующими параметрами:   
- min_samples_split=6 (минимум объектов в узле для разделения),  
- min_samples_leaf=3 (минимальное количество объектов в одном листе),  
- max_features=5 (максимум признаков, которые контролируются при поиске решения).   
Метрика SMAPE для тестовой выборки 14.3%, что может говорить о небольшой ошибке предсказаний.
 
Модель линейной регрессии, скорее всего, слишком простая для определения уровня удовлетворенности работой и не учитывает так много факторов, как модель деревьев решений.
 
 Деревья решений хорошо справляются с задачами регрессии, особенно когда признаки имеют сложные и нелинейные взаимосвязи, что часто характерно для предсказания удовлетворенности работой. Факторы, влияющие на удовлетворенность, могут взаимодействовать нелинейно, и модель дерева решений может адаптироваться к этим сложным отношениям, не требуя дополнительных предположений о распределении данных.
 

В результате анализа SHAP значений выявлено:   
Самой важной составляющей высокой удовлетворенности сотрудника работой является высокая оценка руководителя.  
Количество лет в компании также увеличивает оценку сотрудников, возможно, сотрудники чувствуют себя увереннее, спокойнее, а компания предоставляет пропорциональные стажу бонусы.  
Высокий позитивный вклад в удовлетворенность есть у сотрудников уровня джуниор.




**Для второй задачи прогнозирования увольнения сотрудника:**

Подготовка данных и обучение лучшей модели реализовано в pipe_final_selection. Гиперпараметры моделей деревьев решений и линейной регрессии указаны в словаре parameters_grid. Для оценки качества моделей использована метрика ROC-AUC, так как компании одинаково важно определить оба класса сотрудников.  

Подбор лучшей модели осуществлялся с помощью рандомизированного поиска. В результате рандомизированного поиска выявлена лучшая модель - DecisionTreeClassifier со следующими параметрами:  
- min_samples_split=11 (минимум объектов в узле для разделения),
- min_samples_leaf=3 (минимальное количество объектов в одном листе),
- max_features=4 (максимум признаков, которые контролируются при поиске решения).

Ограничение этих параметров помогает снизить вероятность переобучения модели. Однако, для лучшего результата использован предварительный отбор признаков методом SelectKBest, который позволил увеличить метрику ROC-AUC до 0.92 на тестовых данных.
Самыми важными признаками для предсказания увольнения сотрудника оказались длительность трудоустройства и отделы сотрудников, зарплата.

Деревья решений хорошо справляются с задачами классификации, где важна интерпретируемость и где признаки могут иметь сложные нелинейные зависимости. Для задачи прогнозирования увольнений это важно, так как вероятные причины увольнения могут зависеть от сочетания многих факторов.
Дерево решений разделяет пространство признаков на несколько областей, что позволяет хорошо адаптироваться к структуре данных, не требуя дополнительных сложных предположений о распределении данных.
